```python
import numpy as np
import matplotlib.pyplot as plt

# Las fotos de la clase estan en las fotos del 12 de agosto
```

```python
Delta = 0.1 # Delta = Ts = 1/fs
```

```python
t1 = np.arange(-10, 0, Delta)
t2 = np.arange(0, 10+Delta, Delta)
t = np.concatenate((t1, t2))
```

```python
x1_t = np.abs(t1)
x2_t = t2
```

```python
x_t = np.concatenate((x1_t, x2_t))
```

```python
plt.plot(t, x_t)


plt.show()
```

```python
t1r = -t1[::-1]
t2r = -t2[::-1]
tr = np.concatenate((t2r, t1r))
```

```python
x1_tr = x1_t
x2_tr = x2_t

x_tr = np.concatenate((x1_tr, x2_tr))
```

```python
xp_t = (x_t + x_tr)/2
xi_t = (x_t - x_tr)/2
```

```python
plt.plot(t, xp_t)
plt.show()
plt.plot(t, xi_t)
plt.show()
```

```python
xt = xp_t + xi_t
```

```python
plt.plot(t, xt)
plt.show()
```
