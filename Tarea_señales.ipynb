```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Delta = 0.1
```

```python
t1 = np.arange(-10, 0, Delta)
t2 = np.arange(0, 10+Delta, Delta)
t = np.concatenate((t1, t2))
```

```python
x1_t = -10*np.ones_like(t1)
x2_t = t2
x_t = np.concatenate((x1_t, x2_t))
```

```python
plt.plot(t, x_t)
plt.show()
```

```python
t1r = t1[::-1]
t2r = t2[::-1]
tr = np.concatenate((t1r, t2r))
```

```python
x1_tr = x1_t[::-1]
x2_tr = x2_t[::-1]
xr_tr = np.concatenate((x1_tr, x2_tr))
```

```python
xp_t = (x_t + xr_tr)/2
xi_t = (x_t - xr_tr)/2
```

```python
plt.plot(t, xp_t)
plt.show()
```

```python
plt.plot(t, xi_t)
plt.show()
```

```python
xsuma = xp_t + xi_t
```

```python
plt.plot(t, xsuma)
plt.show()
```
