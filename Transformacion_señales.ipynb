```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1 = -4
p2 = -2
p3 = 0
p4 = 2
p5 = 4
```

```python
#Crear vectores de tiempo
Delta = 0.01
t1 = np.arange(p1, p2, Delta) #p2-Delta
t2 = np.arange(p2, p3, Delta)
t3 = np.arange(p3, p4, Delta)
t4 = np.arange(p4, p5+Delta, Delta)
```

```python
x1 = t1+4
x2 = 2*np.ones(len(t2))
x3 = np.ones(len(t3))
x4 = 2 - t4/2
```

```python
x_t = np.concatenate((x1, x2, x3, x4))
t = np.concatenate((t1, t2, t3, t4))
```

```python
plt.plot(t, x_t)
plt.grid()
plt.show()
```

```python
plt.stem(t, x_t) #Para graficas discretas, se cambia el delta para poder observar
plt.grid()
plt.show()
```

Desplazamiento en el tiempo x(t) --> x(t-2)

```python
t0 = 2 #Retardo = 2
p1r = -4 + t0
p2r = -2 + t0
p3r = 0 + t0
p4r = 2 + t0
p5r = 4 + t0
```

```python
t1r = np.arange(p1r, p2r, Delta) #p2-Delta
t2r = np.arange(p2r, p3r, Delta)
t3r = np.arange(p3r, p4r, Delta)
t4r = np.arange(p4r, p5r+Delta, Delta)
```

```python
x_tr = x_t
tr = np.concatenate((t1r, t2r, t3r, t4r))
```

```python
plt.plot(tr, x_tr)
plt.grid()
plt.show()
```

Escalamiento en el tiempo x(t) --> x(2t - 2)

```python
esc = 2
tesc = np.concatenate((t1r/esc, t2r/esc, t3r/esc, t4r/esc))
```

```python
x_tre = x_t
```

```python
plt.plot(tesc, x_tre)
plt.grid()
plt.show()
```
