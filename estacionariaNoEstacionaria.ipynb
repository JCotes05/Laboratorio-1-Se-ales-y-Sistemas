```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 20   # T = 1/20 = 0.05s
f2 = 40   # T = 1/40 = 0.025s
f3 = 60   # T = 1/60 = 0.01666s
fs = 10*f3  # fs = 600Hz
Delta = 1/fs # Delta = Ts = 1/fs
```

```python
t = np.arange(0, 0.1+Delta, Delta)  # Vector de tiempo para las graficas
x1_t = np.sin(2*np.pi*f1*t)
x2_t = np.sin(2*np.pi*f2*t)
x3_t = np.sin(2*np.pi*f3*t)
```

```python
x_t = x1_t + x2_t + x3_t
```

```python
plt.subplot(2, 2, 1)  # Divide en: Numero de filas, Numero de columnas, Posicion
plt.plot(t, x1_t)
plt.subplot(2, 2, 2)
plt.plot(t, x2_t)
plt.subplot(2, 2, 3)
plt.plot(t, x3_t)
plt.subplot(2, 2, 4)
plt.plot(t, x_t)
plt.show()
```

```python
xne_t = np.concatenate((x1_t, x2_t, x3_t))  #Unir las tres ondas sinusoidales
tne = np.arange(0, 0.3+3*Delta, Delta)
# Otro metodo:
t1 = np.arange(0, 0.1+Delta, Delta)
t2 = np.arange(0.1+Delta, 0.2+2*Delta, Delta)
t3 = np.arange(0.2+2*Delta, 0.3+3*Delta, Delta)
tc = np.concatenate((t1, t2, t3))
```

```python
plt.subplot(1, 2, 1)
plt.plot(tne, xne_t)
plt.subplot(1, 2, 2)
plt.plot(tc, xne_t)
plt.show()
```
