```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_in = -2
n_fin = 3
n = np.arange(n_in, n_fin+1)
x_n = [1, 6, 2, 7, 3, 2]
```

```python
M = 1/3
Z = int(1/M)
L_n = len(x_n)
n_I = np.arange(n_in*1/M, n_fin*1/M + 1)
L_nI = len(n_I)
x_nI = np.arange(1, L_nI+1, dtype=float) #Vector para interpolacion lineal
```

```python
k = 0
for i in range(L_n-1):
  x_nI[k] = x_n[i]
  for j in range(1, Z):
    Dif = x_n[i+1]-x_n[i] #Operacion para interpolacion
    A = j*M*Dif+x_n[i]
    x_nI[k+1] = A
    k = k+1
  k=k+1
x_nI[L_nI-1] = x_n[L_n-1]
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.stem(n, x_n) #Graficar la señal original
plt.subplot(1, 2, 2)
plt.stem(n_I, x_nI) #Graficar la señal interpolada
plt.show()
```
