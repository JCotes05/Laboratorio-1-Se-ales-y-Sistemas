```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
n_in = 0 #Punto inicial del vector
n_fin = 3 #Punto final del vector
n = np.arange(n_in, n_fin+1)
```

```python
x_n = [1, 3, 4, 2]
M = 1/3 #Factor de esaclamiento: Interpolacion
Z = int(1/M) #1/M
L_n = len(x_n) #Longitud del vector original
nI = np.arange(n_in*1/M, n_fin*1/M + 1)
L_nI = len(nI) #Longitud del vector interpolado
x_nI0 = np.arange(1, L_nI+1) #Vector para interpolacion por cero
x_nIEsc = np.arange(1, L_nI+1) #Vector para interpolacion por escalon
```

```python
for k in range(L_nI):
  if k % Z == 0:
    r = int(k*M)
    x_nI0[k] = x_n[r]
    x_nIEsc[k] = x_n[r]
  else:
    x_nI0[k] = 0 #Interpolacion por cero
    x_nIEsc[k] = x_nIEsc[k-1] #Interpolacion escalon
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.stem(n, x_n) #Graficar la señal original
plt.subplot(1, 3, 2)
plt.stem(nI, x_nI0) #Graficar la señal interpolada por cero
plt.subplot(1, 3, 3)
plt.stem(nI, x_nIEsc)
plt.show()
```
