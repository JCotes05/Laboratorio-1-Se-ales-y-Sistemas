```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 20 #Frecuencia de la se√±al T = 1/10 = 0.1s
f2 = 50
f3 = 70
fs = 10*f3 #Frecuencia de muestreo
Delta = 1/fs #Delta = Ts = 1/fs Tiempo de muestreo


```

```python
t = np.arange(0, 0.3+Delta, Delta)
#t1 = np.arange(0, 0.15+Delta, Delta)
#t2 = np.arange(0, 0.06+Delta, Delta)
#t3 = np.arange(0, 0.042+Delta, Delta)

w1 = 2*np.pi*f1 # w = omega
w2 = 2*np.pi*f2
w3 = 2*np.pi*f3
x1_t = np.cos(w1*t)
x2_t = np.sin(w2*t)
x3_t = np.cos(w3*t)
```

```python
plt.plot(t, x1_t)
plt.axis([0, 0.15, -1.05, 1.05])
plt.show()
plt.plot(t, x2_t)
plt.axis([0, 0.06, -1.05, 1.05])
plt.show()
plt.plot(t, x3_t)
plt.axis([0, 0.042, -1.05, 1.05])
plt.show()
```
